# Training

```python
np.random.seed(0)

def train_data(xn):
    w = np.random.randn(2)
    epoch = 100
    print(w)
    avg_err = []
    
    while(epoch):
        err = []
        for i in range(len(xn)):
            
            xi = xn.iat[i, 0]
            yn = xn.iat[i, 1]
            
            sn = np.dot(w.T, [1, xi])
            
            error = 0.5*(yn-sn)**2
            err.append(error)

            eta = 0.001
            w = w + eta*(yn-sn)*xi

        avg_error = sum(err)/(2*len(err))
        avg_err.append(avg_error)
        epoch -= 1
    
    print(avg_error, w)
    return w
    
w = train_data(training)
```

# Testing

```python
def test(xn, w):
    
    y = []
    for i in range(len(xn)):
        y.append(np.dot(w.T, [1, xn.iat[i, 0]]))
    
    return y

y = test(validation, w)
```
